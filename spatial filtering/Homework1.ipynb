# Median Filter

Applies a median filter to an image. The median filter is defined as the median of all pixels within a local region of an image.

The median filter is normally used to reduce salt and pepper noise in an image, somewhat like the mean filter. However, it often does a better job than the mean filter of preserving useful detail in the image.

```python
# You Answer
import numpy as np
import cv2
import matplotlib.pyplot as plt
def median_filter(img, filterSize = (3, 3), stride = 1, padding = 1):
    new_img = np.zeros((int((img.shape[0]-filterSize[0]+2*padding) / stride + 1) , int((img.shape[1]-filterSize[1]+2*padding) / stride + 1)))
    padded_img = np.zeros(np.array(img.shape)[:2] + padding * 2)
    padded_img[padding:-padding:1, padding:-padding:1] = img
    for i in range(0, new_img.shape[0]):
        for j in range(0, new_img.shape[1]):
            new_img[i, j] = np.median(padded_img[i*stride:filterSize[0] + i * stride, j*stride:filterSize[1] + j*stride])
    return new_img


fig, ax = plt.subplots(3, 3, figsize = (9, 12))
for i in range(3):
    img = cv2.imread("a"+str(i+1)+".jpg", cv2.IMREAD_GRAYSCALE)
    new_img = median_filter(img)
    new_img2 = median_filter(img, filterSize=(5, 5), padding=2)

    fig.set_facecolor("white")
    ax[i][0].imshow(img, 'gray');
    ax[i][0].title.set_text("original")
    ax[i][0].axis("off")
    ax[i][1].imshow(new_img, 'gray');
    ax[i][1].title.set_text("3x3 filter applied")
    ax[i][1].axis("off")
    ax[i][2].imshow(new_img2, 'gray');
    ax[i][2].title.set_text("5x5 filter applied")
    ax[i][2].axis("off")
```

# Arithmetic Mean Filter

Applies a arithmetic mean filter to an image.

An arithmetic mean filter operation on an image removes short tailed noise such as uniform and Gaussian type noise from the image at the cost of blurring the image. The arithmetic mean filter is defined as the average of all pixels within a local region of an image.

Pixels that are included in the averaging operation are specified by a mask. The larger the filtering mask becomes the more predominant the blurring becomes and less high spatial frequency detail that remains in the image.

```python
# You Answer
import numpy as np
import cv2
import matplotlib.pyplot as plt
def mean_filter(img, filterSize = (3, 3), stride = 1, padding = 1):
    new_img = np.zeros((int((img.shape[0]-filterSize[0]+2*padding) / stride + 1) , int((img.shape[1]-filterSize[1]+2*padding) / stride + 1)))
    padded_img = np.zeros(np.array(img.shape)[:2] + padding * 2)
    padded_img[padding:-padding:1, padding:-padding:1] = img
    for i in range(0, new_img.shape[0]):
        for j in range(0, new_img.shape[1]):
            new_img[i, j] = np.mean(padded_img[i * stride:filterSize[0] + i * stride, j * stride:filterSize[1] + j * stride])
    return new_img

fig, ax = plt.subplots(3, 3, figsize = (9, 12))
for i in range(3):
    img = cv2.imread("a"+str(i+1)+".jpg", cv2.IMREAD_GRAYSCALE)
    new_img = mean_filter(img)
    new_img2 = mean_filter(img, filterSize=(5, 5), padding=2)

    fig.set_facecolor("white")
    ax[i][0].imshow(img, 'gray');
    ax[i][0].title.set_text("original")
    ax[i][0].axis("off")
    ax[i][1].imshow(new_img, 'gray');
    ax[i][1].title.set_text("3x3 filter applied")
    ax[i][1].axis("off")
    ax[i][2].imshow(new_img2, 'gray');
    ax[i][2].title.set_text("5x5 filter applied")
    ax[i][2].axis("off")
```

# Geometric Mean Filter

Applies a geometric mean filter to an image.

In the geometric mean method, the color value of each pixel is replaced with the geometric mean of color values of the pixels in a surrounding region. 

The geometric mean filter is better at removing Gaussian type noise and preserving edge features than the arithmetic mean filter. The geometric mean filter is very susceptible to negative outliers.

The geometric mean filter is better at removing Gaussian type noise and preserving edge features than the arithmetic mean filter. The geometric mean filter is very susceptible to negative outliers.

```python
# You Answer
import numpy as np
import cv2
import matplotlib.pyplot as plt
def geometric_mean_filter(img, filterSize = (3, 3), stride = 1, padding = 1):
    new_img = np.zeros((int((img.shape[0]-filterSize[0]+2*padding) / stride + 1) , int((img.shape[1]-filterSize[1]+2*padding) / stride + 1)))
    padded_img = np.ones(np.array(img.shape)[:2] + padding * 2)
    padded_img[padding:-padding:1, padding:-padding:1] = img
    for i in range(0, new_img.shape[0]):
        for j in range(0, new_img.shape[1]):
            new_img[i, j] = np.prod(padded_img[i*stride:filterSize[0] + i*stride, j*stride:filterSize[1] + j*stride]) ** (1/(filterSize[0] * filterSize[1]))
    return new_img

fig, ax = plt.subplots(3, 3, figsize = (9, 12))
for i in range(3):
    img = cv2.imread("a"+str(i+1)+".jpg", cv2.IMREAD_GRAYSCALE)
    new_img = geometric_mean_filter(img)
    new_img2 = geometric_mean_filter(img, filterSize=(5, 5), padding=2)

    fig.set_facecolor("white")
    ax[i][0].imshow(img, 'gray');
    ax[i][0].title.set_text("original")
    ax[i][0].axis("off")
    ax[i][1].imshow(new_img, 'gray');
    ax[i][1].title.set_text("3x3 filter applied")
    ax[i][1].axis("off")
    ax[i][2].imshow(new_img2, 'gray');
    ax[i][2].title.set_text("5x5 filter applied")
    ax[i][2].axis("off")
```

# Harmonic Mean Filter

Applies a harmonic mean filter to an image.

In the harmonic mean method, the color value of each pixel is replaced with the harmonic mean of color values of the pixels in a surrounding region.
The harmonic mean filter is better at removing Gaussian type noise and preserving edge features than the arithmetic mean filter. The harmonic mean filter is very good at removing positive outliers.

```python
# You Answer
import numpy as np
import cv2
import matplotlib.pyplot as plt
def harmonic_mean_filter(img, filterSize = (3, 3), stride = 1, padding = 1):
    new_img = np.zeros((int((img.shape[0]-filterSize[0]+2*padding) / stride + 1) , int((img.shape[1]-filterSize[1]+2*padding) / stride + 1)))
    padded_img = np.ones(np.array(img.shape)[:2] + padding * 2)
    padded_img[padding:-padding:1, padding:-padding:1] = img
    eps = 1e-10
    for i in range(0, new_img.shape[0]):
        for j in range(0, new_img.shape[1]):
            new_img[i, j] = np.prod(filterSize) / np.sum((eps+padded_img[i*stride:filterSize[0] + i*stride, j*stride:filterSize[1] + j*stride]) ** (-1))
    return new_img

fig, ax = plt.subplots(3, 3, figsize = (9, 12))
for i in range(3):
    img = cv2.imread("a"+str(i+1)+".jpg", cv2.IMREAD_GRAYSCALE)
    new_img = harmonic_mean_filter(img)
    new_img2 = harmonic_mean_filter(img, filterSize=(5, 5), padding=2)

    fig.set_facecolor("white")
    ax[i][0].imshow(img, 'gray');
    ax[i][0].title.set_text("original")
    ax[i][0].axis("off")
    ax[i][1].imshow(new_img, 'gray');
    ax[i][1].title.set_text("3x3 filter applied")
    ax[i][1].axis("off")
    ax[i][2].imshow(new_img2, 'gray');
    ax[i][2].title.set_text("5x5 filter applied")
    ax[i][2].axis("off")
```

# Contraharmonic Mean Filter

Applies a contraharmonic mean filter to an image.

With a contraharmonic mean filter, the color value of each pixel is replaced with the contraharmonic mean of color values of the pixels in a surrounding region.

A contraharmonic mean filter reduces or virtually eliminates the effects of salt-and-pepper noise. For positive values of Q, the filter eliminates pepper noise. For negative values of Q it eliminates salt noise. It cannot do both simultaneously.

Note that the contraharmonic filter is simply the arithmetic mean filter if Q = 0, and the harmonic mean filter if Q = 1.

```python
# You Answer
import numpy as np
import cv2
import matplotlib.pyplot as plt
def  contraharmonic_mean_filter(img, filterSize = (3, 3), stride = 1, padding = 1, q = 1):
    new_img = np.zeros((int((img.shape[0]-filterSize[0]+2*padding) / stride + 1) , int((img.shape[1]-filterSize[1]+2*padding) / stride + 1)))
    padded_img = np.zeros(np.array(img.shape)[:2] + padding * 2)
    padded_img[padding:-padding:1, padding:-padding:1] = img
    eps = 1e-10
    for i in range(0, new_img.shape[0]):
        for j in range(0, new_img.shape[1]):
            new_img[i, j] = np.sum(padded_img[i*stride:filterSize[0] + i*stride, j*stride:filterSize[1] + j*stride]) ** (q + 1) / np.sum((eps+padded_img[i*stride:filterSize[0] + i*stride, j*stride:filterSize[1] + j*stride]) ** q)
    return new_img

fig, ax = plt.subplots(3, 4, figsize = (12, 12))
for i in range(3):
    img = cv2.imread("a"+str(i+1)+".jpg", cv2.IMREAD_GRAYSCALE)
    new_img = contraharmonic_mean_filter(img)
    new_img2 = contraharmonic_mean_filter(img, filterSize=(3, 3), q = 1)
    new_img3 = contraharmonic_mean_filter(img, filterSize=(3, 3), q = -1)

    fig.set_facecolor("white")
    ax[i][0].imshow(img, 'gray');
    ax[i][0].title.set_text("original")
    ax[i][0].axis("off")
    ax[i][1].imshow(new_img, 'gray');
    ax[i][1].title.set_text("3x3, q = 1 filter applied")
    ax[i][1].axis("off")
    ax[i][2].imshow(new_img2, 'gray');
    ax[i][2].title.set_text("3x3, q = 2 filter applied")
    ax[i][2].axis("off")
    ax[i][3].imshow(new_img3, 'gray');
    ax[i][3].title.set_text("3x3, q = -1 filter applied")
    ax[i][3].axis("off")
```

# Maximum Filter

Applies a maximum filter to an image. The maximum filter is defined as the maximum of all pixels within a local region of an image.

The maximum filter is typically applied to an image to remove negative outlier noise.

```python
# You Answer
import numpy as np
import cv2
import matplotlib.pyplot as plt
def max_filter(img, filterSize = (3, 3), stride = 1, padding = 1):
    new_img = np.zeros((int((img.shape[0]-filterSize[0]+2*padding) / stride + 1) , int((img.shape[1]-filterSize[1]+2*padding) / stride + 1)))
    padded_img = np.zeros(np.array(img.shape)[:2] + padding * 2)
    padded_img[padding:-padding:1, padding:-padding:1] = img
    for i in range(0, new_img.shape[0]):
        for j in range(0, new_img.shape[1]):
            new_img[i, j] = np.max(padded_img[i*stride:filterSize[0] + i*stride, j*stride:filterSize[1] + j*stride], axis = None)
    return new_img

fig, ax = plt.subplots(3, 3, figsize = (9, 12))
for i in range(3):
    img = cv2.imread("a"+str(i+1)+".jpg", cv2.IMREAD_GRAYSCALE)
    new_img = max_filter(img)
    new_img2 = max_filter(img, filterSize=(5, 5), padding=2)

    fig.set_facecolor("white")
    ax[i][0].imshow(img, 'gray');
    ax[i][0].title.set_text("original")
    ax[i][0].axis("off")
    ax[i][1].imshow(new_img, 'gray');
    ax[i][1].title.set_text("3x3 filter applied")
    ax[i][1].axis("off")
    ax[i][2].imshow(new_img2, 'gray');
    ax[i][2].title.set_text("5x5 filter applied")
    ax[i][2].axis("off")
```

# Midpoint Filter

Applies a midpoint filter to an image.

In the midpoint method, the color value of each pixel is replaced with the average of maximum and minimum (i.e. the midpoint) of color values of the pixels in a surrounding region. A larger region (filter size) yields a stronger effect.

The midpoint filter is typically used to filter images containing short tailed noise such as Gaussian and uniform type noise.

```python
# You Answer
import numpy as np
import cv2
import matplotlib.pyplot as plt
def midpoint_filter(img, filterSize = (3, 3), stride = 1, padding = 1):
    new_img = np.zeros((int((img.shape[0]-filterSize[0]+2*padding) / stride + 1) , int((img.shape[1]-filterSize[1]+2*padding) / stride + 1)))
    padded_img = np.zeros(np.array(img.shape)[:2] + padding * 2)
    padded_img[padding:-padding:1, padding:-padding:1] = img
    for i in range(0, new_img.shape[0]):
        for j in range(0, new_img.shape[1]):
            new_img[i, j] = int((np.max(padded_img[i*stride:filterSize[0] + i*stride, j*stride:filterSize[1] + j*stride], axis = None) + \
                                       np.min(padded_img[i*stride:filterSize[0] + i*stride, j*stride:filterSize[1] + j*stride], axis = None)) / 2)
    return new_img

fig, ax = plt.subplots(3, 3, figsize = (9, 12))
for i in range(3):
    img = cv2.imread("a"+str(i+1)+".jpg", cv2.IMREAD_GRAYSCALE)
    new_img = midpoint_filter(img)
    new_img2 = midpoint_filter(img, filterSize=(5, 5), padding=2)

    fig.set_facecolor("white")
    ax[i][0].imshow(img, 'gray');
    ax[i][0].title.set_text("original")
    ax[i][0].axis("off")
    ax[i][1].imshow(new_img, 'gray');
    ax[i][1].title.set_text("3x3 filter applied")
    ax[i][1].axis("off")
    ax[i][2].imshow(new_img2, 'gray');
    ax[i][2].title.set_text("5x5 filter applied")
    ax[i][2].axis("off")
```

# Range Filter

Applies a Range filter to an image.

In the range method, the color value of each pixel is replaced with the difference of maximum and minimum (i.e. the range) of color values of the pixels in a surrounding region.

The range filter is used to find edges within an image.

```python
# You Answer
import numpy as np
import cv2
import matplotlib.pyplot as plt
def range_filter(img, filterSize = (3, 3), stride = 1, padding = 1):
    new_img = np.zeros((int((img.shape[0]-filterSize[0]+2*padding) / stride + 1) , int((img.shape[1]-filterSize[1]+2*padding) / stride + 1)))
    padded_img = np.zeros(np.array(img.shape)[:2] + padding * 2)
    padded_img[padding:-padding:1, padding:-padding:1] = img
    for i in range(0, new_img.shape[0], stride):
        for j in range(0, new_img.shape[1], stride):
            new_img[i, j] = int((np.max(padded_img[i:filterSize[0] + i, j:filterSize[1] + j], axis = None) - \
                                       np.min(padded_img[i:filterSize[0] + i, j:filterSize[1] + j], axis = None)))
    return new_img

fig, ax = plt.subplots(3, 3, figsize = (9, 12))
for i in range(3):
    img = cv2.imread("a"+str(i+1)+".jpg", cv2.IMREAD_GRAYSCALE)
    new_img = range_filter(img)
    new_img2 = range_filter(img, filterSize=(5, 5), padding=2)

    fig.set_facecolor("white")
    ax[i][0].imshow(img, 'gray');
    ax[i][0].title.set_text("original")
    ax[i][0].axis("off")
    ax[i][1].imshow(new_img, 'gray');
    ax[i][1].title.set_text("3x3 filter applied")
    ax[i][1].axis("off")
    ax[i][2].imshow(new_img2, 'gray');
    ax[i][2].title.set_text("5x5 filter applied")
    ax[i][2].axis("off")
```

# Emboss Filter

Generates an embossed image.

Image embossing is a computer graphics technique in which each pixel of an image is replaced either by a highlight or a shadow, depending on light/dark boundaries on the original image. Low contrast areas are replaced by a gray background.

```python
# You Answer
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
def emboss_filter(img, filter, padding = 1):
    padded_img = np.zeros(np.array(img.shape)[:2] + padding * 2)
    padded_img[padding:-padding:1, padding:-padding:1] = img
    new_img = convolve2d(padded_img, filter)
    return new_img

fig, ax = plt.subplots(3, 4, figsize = (12, 12))
filter1 = np.array([[0, 1, 0], [0, 0, 0], [0, -1, 0]])
filter2 = np.array([[0, 0, 0], [1, 0, -1], [0, 0, 0]])
filter3 = np.array([[1, 1, 0], [1, 0, -1], [0, -1, -1]])
for i in range(3):
    img = cv2.imread("a"+str(i+1)+".jpg", cv2.IMREAD_GRAYSCALE)
    new_img = emboss_filter(img, filter1)
    new_img2 = emboss_filter(img, filter2)
    new_img3 = emboss_filter(img, filter3)

    fig.set_facecolor("white")
    ax[i][0].imshow(img, 'gray');
    ax[i][0].title.set_text("original")
    ax[i][0].axis("off")
    ax[i][1].imshow(new_img, 'gray');
    ax[i][1].title.set_text("filter1 applied")
    ax[i][1].axis("off")
    ax[i][2].imshow(new_img2, 'gray');
    ax[i][2].title.set_text("filter2 applied")
    ax[i][2].axis("off")
    ax[i][3].imshow(new_img3, 'gray');
    ax[i][3].title.set_text("filter3 applied")
    ax[i][3].axis("off")
```
